```python
import cycler

# Defines a cycle of colors using the cycler to use as a cycle for different plot elements

colors = cycler.cycler('color', ['green', 'yellow', 'lawngreen', 'gold', 'greenyellow', 'goldenrod'])         
"""
    cycler.cycler is a way to access the cycler function within the cycler module
"""

#colors = cycler('color', ['#32CD32', '#FFFF00', '#7CFC00', '#FFD700', '#ADFF2F', '#DAA520'])         # green and yellow


#colors = cycler('color', ['#FF6347', '#20B2AA', '#9370DB', '#3CB371', '#FFA07A', '#4682B4'])        # red and light blue
#colors = cycler('color', ['#FF5733', '#33FF57', '#3357FF', '#FF33A1', '#33A1FF', '#FFD700'])         # red and green
#colors = cycler('color', ['#008000', '#FFA500', '#1E90FF', '#8A2BE2', '#FF4500', '#2E8B57'])



plt.rc('axes', facecolor="#333333", edgecolor='none', axisbelow=True, grid=True, prop_cycle=colors, labelcolor='dimgray')
"""
    facecolor="#333333": This sets the background color of the axes to a dark gray color with the hexadecimal code #333333

    edgecolor='none': This removes the edge color of the axes, making it transparent.            it can be for example "black"

    axisbelow=True: This ensures that the grid and data are plotted below the other plot elements.

    grid=False: This disables the grid lines on the plot.

    prop_cycle=colors: This sets the color cycle for multiple lines or markers in the plot. 
                       The colors variable should be a list of colors to cycle through.

    labelcolor='gray': This sets the color of the axis labels to gray.
"""

plt.rc('figure', facecolor='#333333')


plt.rc('grid', color='6c7171', linestyle='--') 
"""
    'solid' or '-' : Solid line, 
    'dashed' or '--' : Dashed line, 
    'dashdot' or '-.' : Dash-dot line, 
    'dotted' or ':' : Dotted line
"""

plt.rc('xtick', color='6c7171')
plt.rc('ytick', color='6c7171')
plt.rc("text", color="white")       # for title and legend
```
